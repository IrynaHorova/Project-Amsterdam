# Sports & Culture impact on employment

```python
grants_empl = data
```

```python
grants_empl['average_grant_amount_total'] = grants_empl['amount_given_Total'] / grants_empl['nr_subsidies_Total']
grants_empl['average_grant_amount_Sport'] = grants_empl['amount_given_Sport'] / grants_empl['nr_subsidies_Sport']
grants_empl['average_grant_amount_Culture'] = grants_empl['amount_given_Culture'] / grants_empl['nr_subsidies_Culture']
```

```python
selected_grants_empl = grants_empl[['year', 
                                    'amount_given_Total',  
                                    'nr_subsidies_Total',
                                    'average_grant_amount_total',
                                    'startups_in_sport_culture',
                                    'employers_sports_culture',
                                    'pct_employers_sports_cultures',
                                    'employees_sports_culture',
                                    'pct_employees_sports_culture'
                                   ]]\
.set_index('year')

selected_grants_empl.transpose()
```

```python
selected_grants_empl.reset_index()['year']
```

# percentage growth

```python
grants_empl_percent_change = selected_grants_empl.pct_change()
grants_empl_percent_change['year'] = selected_grants_empl.reset_index()['year']
pd.options.display.float_format = '{:,.2f}'.format
grants_empl_percent_change.transpose()
```

# correlations

```python
df1 = selected_grants_empl.loc[:, selected_grants_empl.columns != 'startups_in_sport_culture']
df2 = selected_grants_empl.loc[:, selected_grants_empl.columns != 'startups_in_sport_culture']

coeffmat = np.zeros((df1.shape[1], df2.shape[1]))
pvalmat = np.zeros((df1.shape[1], df2.shape[1]))

for i in range(df1.shape[1]):    
    for j in range(df2.shape[1]):        
        corrtest = pearsonr(df1[df1.columns[i]], df2[df2.columns[j]])  
        coeffmat[i,j] = corrtest[0]
        pvalmat[i,j] = corrtest[1]

# dfcoeff = pd.DataFrame(coeffmat, columns=df2.columns, index=df1.columns)
# dfcoeff
# #           Col1      Col2      Col3      Col4      Col5
# # Col1 -0.791083  0.459101 -0.488463 -0.289265  0.494897
# # Col2  0.059446 -0.395072  0.310900  0.297532  0.201669
# # Col3 -0.062592  0.391469 -0.450600 -0.136554  0.299579
# # Col4 -0.470203  0.797971 -0.193561 -0.338896 -0.244132
# # Col5 -0.057848 -0.037053  0.042798  0.176966 -0.157344

dfpvals = pd.DataFrame(pvalmat, columns=df2.columns, index=df1.columns)
print('correlation significance')
dfpvals
#           Col1      Col2      Col3      Col4      Col5
# Col1  0.006421  0.181967  0.152007  0.417574  0.145871
# Col2  0.870421  0.258506  0.381919  0.403770  0.576357
# Col3  0.863615  0.263268  0.191245  0.706796  0.400385
# Col4  0.170260  0.005666  0.592096  0.338101  0.496668
# Col5  0.873881  0.919058  0.906551  0.624783  0.664206
```
