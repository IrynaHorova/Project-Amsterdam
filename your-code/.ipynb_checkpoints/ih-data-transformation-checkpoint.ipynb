```python
import pandas as pd
import numpy as np
import xlrd
```

# Grants

```python
grants = pd.read_csv('../data/raw files/Grants-for-sports-etc.csv', sep = ";", encoding = 'ISO-8859-1', na_values = 0)
grants = grants[1:].reset_index()
grants = grants[:-2]
```

```python
translate_columns ={
    "TABLE_NAME" : "table_name",
    "DOSSIERNUMMER" : "dossier_numbers",
    "AANVRAGER" : "applicant",
    "REGELINGNAAM" : "regulation_name",
    "BELEIDSTERREIN" : "policy_area",
    "ORGANISATIEONDERDEEL" : "organisational_part",
    "PROJECT_NAAM" : "project_name",
    "TYPE_PERIODICITEIT" : "type_periodicity",
    "BEDRAG_AANGEVRAAGD" : "amount_requested",
    "BEDRAG_VERLEEND" : "amount_granted",
    "BEDRAG_VASTGESTELD" : "amount_given",
    "SUBSIDIEJAAR" : "year",
    "DATUM_OVERZICHT" : "date_overview"
    
}

grants = grants.rename(columns = translate_columns)

grants.head()
```

```python
#grants['policy_area'].unique()

translation_policy_area = {
    'Welzijn en Zorg' : 'Welfare and Care', 
    'Algemeen':"General", 
    'Cultuur':"Culture", 
    'Onderwijs':"Education",
    'Participatie':"Particiation", 
    'Sport':"Sport", 
    'Jeugd':"Youth", 
    'Inkomen':"Income",
    'Openbare orde en veiligheid':"Public order and safety", 
    'Verkeer en openbare ruimte':"Traffic and public space",
    'Stedelijke ontwikkeling en wonen':"Urban development and living", 
    'Bestuur en ondersteuning':"Governance and support",
    'Economie':"Economy", 
    'Ruimte en Duurzaamheid':"Space and Sustainability", 
    'Werk':"Work", 
    'Zorg':"Care", 
    'Monumenten':"Monuments",
    'Sociale basis':"Social basis"
}
```

```python
grants['policy_area_eng'] = grants['policy_area'].map(translation_policy_area)
grants = grants.drop('index', axis = 1 )
grants[["amount_requested", "amount_granted", "amount_given", "year"]] = \
grants[["amount_requested", "amount_granted", "amount_given", "year"]].apply(pd.to_numeric)
grants.dtypes
```

```python
grants1 = grants[["year", "organisational_part", "type_periodicity", "policy_area_eng", "amount_requested", "amount_granted", "amount_given"]]
grants2 = grants1[grants1["policy_area_eng"].isin(['Sport','Culture'])]
grants2 = grants2.fillna(0)
grants2.head()
```

```python
res = grants2[grants2['year'] == 2016].describe().transpose()
desc = pd.DataFrame(columns = res.columns)

for ar in ['Culture' , 'Sport']:
    for yr in grants2['year'].unique():
        res = grants2[(grants2['year'] == yr) & (grants2['policy_area_eng'] == ar)].describe().transpose()
        res['year'] = yr
        res['area'] = ar
        desc = desc.append(res)

desc = desc.drop('year', axis = 0 )
desc = desc[desc['year'] <= 2020]
desc.to_csv('../data/cleansed/cle_grants_employment_descriptives.csv', index = False)
```

```python
grants3 = pd.get_dummies(grants2['type_periodicity'])
grants4 = pd.concat([grants2[['year', 'organisational_part', 'policy_area_eng', 'amount_requested', 'amount_granted','amount_given']], grants3], axis = 1)
grants4.loc[(grants4['amount_given'] == 0), 'Eenmalig']= 0
grants4.loc[(grants4['amount_given'] == 0), 'Periodiek'] = 0

grants4.head()
```

```python
def count_positive(x):
    return np.sum(np.array(x) > 0)
```

```python
grants4['amount_given_copy'] = grants4['amount_given']
```

```python
grants5 = grants4.groupby(
    ['year','policy_area_eng'])["amount_requested", "amount_granted", "amount_given", "amount_given_copy",'Eenmalig', 'Periodiek']\
    .agg({"amount_requested":['sum'], "amount_granted":['sum'], 'amount_given':['sum'], 'Eenmalig':['sum'], 'Periodiek':['sum'], 'amount_given_copy':[count_positive]}
).reset_index()

grants5 = grants5[(grants5['year'] <= 2020)]
grants5.columns = grants5.columns.map('_'.join)

grants5
```

```python
grants5 = grants5.pivot(index="year_", columns="policy_area_eng_", values=["amount_requested_sum", "amount_granted_sum", "amount_given_sum", 'Eenmalig_sum', 'Periodiek_sum', 'amount_given_copy_count_positive']).reset_index()
grants5.columns = grants5.columns.map('_'.join)
grants5
```

```python
grants5 = grants5.rename(columns = {'year__':"year", 
                                    'amount_requested_sum_Culture':"amount_requested_Culture",
                                    'amount_requested_sum_Sport':"amount_requested_Sport",
                                    'amount_granted_sum_Culture':"amount_granted_Culture",
                                    'amount_granted_sum_Sport':"amount_granted_Sport",
                                    'amount_given_sum_Culture':"amount_given_Culture",
                                    'amount_given_sum_Sport':"amount_given_Sport",
                                    'Eenmalig_sum_Culture' : "nr_one_off_Culture",
                                    'Periodiek_sum_Culture' : "nr_periodic_Culture",
                                    'Eenmalig_sum_Sport' : "nr_one_off_Sport",
                                    'Periodiek_sum_Sport' : "nr_periodic_Sport",
                                    'amount_given_copy_count_positive_Culture' : "nr_subsidies_Culture", 
                                    'amount_given_copy_count_positive_Sport' : "nr_subsidies_Sport"
                                   })
grants5 = grants5.fillna(0)

grants5['amount_requested_Total'] = grants5['amount_requested_Sport'] + grants5['amount_requested_Culture']
grants5['amount_granted_Total'] = grants5['amount_granted_Sport'] + grants5['amount_granted_Culture']
grants5['amount_given_Total'] = grants5['amount_given_Sport'] + grants5['amount_given_Culture']
grants5['nr_subsidies_Total'] = grants5['nr_subsidies_Culture'] + grants5['nr_subsidies_Sport']

grants5
```

# startups-in-sports-and-culture

```python
xls = pd.ExcelFile('../data/raw files/#startups-in-sports-and-culture.xlsx')
df = xls.parse('2020_jaarboek_518', skiprows=2)
startups = df.iloc[1:,:]
startups = startups.iloc[:-3,:]
startups = startups.applymap(lambda x: x.replace('-','0') if type(x) is str else x)
startups.head()
```

```python
startups[[2015,2016,2017,2018,2019]] = startups[[2015,2016,2017,2018,2019]].apply(pd.to_numeric)
startups.dtypes
```

```python
startups = startups[startups['omschrijving SBI 2008'] == 'cultuur, sport en recreatie']
startups = startups.rename(columns = {'index':"year", 19:"startupsin "})
startups = startups.iloc[:,2:]
startups = startups.transpose().reset_index()
startups = startups.rename(columns = {'index':"year", 19:"startups_in_sport_culture"})
startups
```

```python
df1 = grants5.merge(startups, how = 'left', left_on='year', right_on='year')
df1
```

# Employers in sports

```python
employed_org = employed_org.rename(columns = {'index':"year", 22:"employers_sports_culture"})
employed_org
```
