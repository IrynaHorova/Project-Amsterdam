# Grants

```python
grants = pd.read_csv('../data/raw files/Grants-for-sports-etc.csv', sep = ";", encoding = 'ISO-8859-1', na_values = 0)
grants.head()
```

```python
grants.head(10)
```

```python
grants = grants[1:].reset_index()
grants = grants[:-2]
grants.head(10)
```

```python
translate_columns ={
    "TABLE_NAME" : "table_name",
    "DOSSIERNUMMER" : "dossier_numbers",
    "AANVRAGER" : "applicant",
    "REGELINGNAAM" : "regulation_name",
    "BELEIDSTERREIN" : "policy_area",
    "ORGANISATIEONDERDEEL" : "organisational_part",
    "PROJECT_NAAM" : "project_name",
    "TYPE_PERIODICITEIT" : "type_periodicity",
    "BEDRAG_AANGEVRAAGD" : "amount_requested",
    "BEDRAG_VERLEEND" : "amount_granted",
    "BEDRAG_VASTGESTELD" : "amount_given",
    "SUBSIDIEJAAR" : "year",
    "DATUM_OVERZICHT" : "date_overview"
    
}

grants = grants.rename(columns = translate_columns)

grants.head()
```

```python
grants['policy_area'].unique()
```

```python
translation_policy_area = {
    'Welzijn en Zorg' : 'Welfare and Care', 
    'Algemeen':"General", 
    'Cultuur':"Culture", 
    'Onderwijs':"Education",
    'Participatie':"Particiation", 
    'Sport':"Sport", 
    'Jeugd':"Youth", 
    'Inkomen':"Income",
    'Openbare orde en veiligheid':"Public order and safety", 
    'Verkeer en openbare ruimte':"Traffic and public space",
    'Stedelijke ontwikkeling en wonen':"Urban development and living", 
    'Bestuur en ondersteuning':"Governance and support",
    'Economie':"Economy", 
    'Ruimte en Duurzaamheid':"Space and Sustainability", 
    'Werk':"Work", 
    'Zorg':"Care", 
    'Monumenten':"Monuments",
    'Sociale basis':"Social basis"
}
```

```python
grants['policy_area_eng'] = grants['policy_area'].map(translation_policy_area)
```

```python
grants.head(10)
```

```python
grants.dtypes
```

```python
grants[["amount_requested", "amount_granted", "amount_given", "year"]] = \
grants[["amount_requested", "amount_granted", "amount_given", "year"]].apply(pd.to_numeric)
```

```python
grants.dtypes
```

```python
grants1 = grants[["year", "policy_area_eng", "amount_requested", "amount_granted", "amount_given"]]
grants1
```

```python
grants2 = grants1[grants1["policy_area_eng"].isin(['Sport','Culture'])]
grants2
```

```python
grants2 = grants2.fillna(0)
grants2
```

```python
grants3 = grants2.groupby(["year",'policy_area_eng']).sum().reset_index()
grants3
```

```python
grants3 = grants3.pivot(index="year", columns="policy_area_eng", values=["amount_requested", "amount_granted", "amount_given"]).reset_index()
```

```python
grants3.columns.map('_'.join)
```

```python
grants3.columns = grants3.columns.map('_'.join)
```

```python
grants3 = grants3.rename(columns = {'year_':"year"})
```

```python
grants3 = grants3.fillna(0)
```

```python
grants3
```

```python
grants3['amount_requested_Total'] = grants3['amount_requested_Sport'] + grants3['amount_requested_Culture']
grants3['amount_granted_Total'] = grants3['amount_granted_Sport'] + grants3['amount_granted_Culture']
grants3['amount_given_Total'] = grants3['amount_given_Sport'] + grants3['amount_given_Culture']
```

```python
grants3 = grants3[(grants3['year'] <= 2020)]
grants3
```

# startups-in-sports-and-culture

```python
xls = pd.ExcelFile('../data/raw files/#startups-in-sports-and-culture.xlsx')
df = xls.parse('2020_jaarboek_518', skiprows=2)
startups = df.iloc[1:,:]
startups = startups.iloc[:-3,:]
startups
```

```python
startups = startups.applymap(lambda x: x.replace('-','0') if type(x) is str else x)
startups
```

```python
startups[[2015,2016,2017,2018,2019]] = startups[[2015,2016,2017,2018,2019]].apply(pd.to_numeric)
startups.dtypes
```

```python
startups = startups[startups['omschrijving SBI 2008'] == 'cultuur, sport en recreatie']
```

```python
startups = startups.iloc[:,2:]
```

```python
startups = startups.transpose().reset_index()
startups
```

```python
startups = startups.rename(columns = {'index':"year", 19:"startups_in_sport_culture"})
```

```python
startups
```

```python
df1 = grants3.merge(startups, how = 'left', left_on='year', right_on='year')
```
