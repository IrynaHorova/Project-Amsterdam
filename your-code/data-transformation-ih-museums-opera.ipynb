# Grants

```python
grants = pd.read_csv('../data/raw files/Grants-for-sports-etc.csv', sep = ";", encoding = 'ISO-8859-1', na_values = 0)
grants = grants[1:-2].reset_index()
grants = grants.drop('index', axis = 1 )
```

```python
translate_columns ={
    "TABLE_NAME" : "table_name",
    "DOSSIERNUMMER" : "dossier_numbers",
    "AANVRAGER" : "applicant",
    "REGELINGNAAM" : "regulation_name",
    "BELEIDSTERREIN" : "policy_area",
    "ORGANISATIEONDERDEEL" : "organisational_part",
    "PROJECT_NAAM" : "project_name",
    "TYPE_PERIODICITEIT" : "type_periodicity",
    "BEDRAG_AANGEVRAAGD" : "amount_requested",
    "BEDRAG_VERLEEND" : "amount_granted",
    "BEDRAG_VASTGESTELD" : "amount_given",
    "SUBSIDIEJAAR" : "year",
    "DATUM_OVERZICHT" : "date_overview"
    
}

grants = grants.rename(columns = translate_columns)

grants.tail()
```

```python
grants.shape
```

```python
#grants['policy_area'].unique()

translation_policy_area = {
    'Welzijn en Zorg' : 'Welfare and Care', 
    'Algemeen':"General", 
    'Cultuur':"Culture", 
    'Onderwijs':"Education",
    'Participatie':"Particiation", 
    'Sport':"Sport", 
    'Jeugd':"Youth", 
    'Inkomen':"Income",
    'Openbare orde en veiligheid':"Public order and safety", 
    'Verkeer en openbare ruimte':"Traffic and public space",
    'Stedelijke ontwikkeling en wonen':"Urban development and living", 
    'Bestuur en ondersteuning':"Governance and support",
    'Economie':"Economy", 
    'Ruimte en Duurzaamheid':"Space and Sustainability", 
    'Werk':"Work", 
    'Zorg':"Care", 
    'Monumenten':"Monuments",
    'Sociale basis':"Social basis"
}
```

```python
grants['policy_area_eng'] = grants['policy_area'].map(translation_policy_area)
grants[["amount_requested", "amount_granted", "amount_given", "year"]] = \
grants[["amount_requested", "amount_granted", "amount_given", "year"]].apply(pd.to_numeric)
grants.dtypes
```

```python
grants2 = grants[grants['applicant'].isin(['Stichting stedelijk museum Amsterdam','stichting Amsterdam Museum','Stichting Nationale Opera & Ballet'])]
```

```python
grants2.shape
```

```python
grants2 = grants2.groupby(['year','applicant']).sum().reset_index().sort_values(by = ['applicant','year'])
grants2
```

```python
grants2 = grants2[grants2['applicant'].isin(['Stichting stedelijk museum Amsterdam','stichting Amsterdam Museum','Stichting Nationale Opera & Ballet'])]
grants2
```

# Opera visitors

```python
opera_vis = pd.read_csv('../data/cleansed/#visitors-opera.csv')
```

```python
opera_vis = opera_vis.iloc[:1,:]
```

```python
opera_vis = opera_vis.transpose().iloc[1:,:]

opera_vis['applicant'] = 'Stichting Nationale Opera & Ballet'

opera_vis = opera_vis.reset_index()

opera_vis = opera_vis.rename(columns = {'index':'year', 0:'visitors'})

opera_vis
```

```python
museums = museums.rename(columns = {'Unnamed: 0':'applicant', 'value':'visitors', 'index':'year'})

museums
```

```python
visitors = museums.append(opera_vis[['year','applicant','visitors']])
```

```python
grants2
```

```python
visitors['year'] = visitors['year'].apply(pd.to_numeric)
```

```python
df1 = grants2.merge(visitors, how = 'left', left_on=['year','applicant'], right_on=['year','applicant'])
df1 = df1[['year','applicant','amount_requested','amount_granted','amount_given','visitors']]
df1
```
