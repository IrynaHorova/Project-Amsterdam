```python
# data_fd = data_fd.transpose()
# data_fd = data_fd.set_index('year')
# data_fd = data_fd.drop(columns=['2016', '2020'])
# np.isinf(data_fd).any()
#data_fd = data_fd.fillna(0)
data_fd = data_fd.reset_index()
data_fd = data_fd[(data_fd['year'] !=2016) & (data_fd['year'] !=2020)].set_index('year')
data_fd
```

```python
df1 = data_fd.loc[:, data_fd.columns]
df2 = data_fd.loc[:, data_fd.columns]

# np.isnan(df1).any()
# np.isnan(df2).any()

coeffmat = np.zeros((df1.shape[1], df2.shape[1]))
pvalmat = np.zeros((df1.shape[1], df2.shape[1]))

for i in range(df1.shape[1]):    
    for j in range(df2.shape[1]):        
        corrtest = pearsonr(df1[df1.columns[i]], df2[df2.columns[j]])  
        coeffmat[i,j] = corrtest[0]
        pvalmat[i,j] = corrtest[1]
        
dfpvals = pd.DataFrame(pvalmat, columns=df2.columns, index=df1.columns)
print('correlation significance')
dfpvals
```
